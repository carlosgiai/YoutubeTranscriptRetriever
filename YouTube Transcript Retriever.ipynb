```python


from youtube_transcript_api import YouTubeTranscriptApi
from urllib.parse import urlparse, parse_qs

def extract_video_id(url):
    parsed_url = urlparse(url)
    
    if parsed_url.netloc == 'youtu.be':
        return parsed_url.path[1:]
    
    if parsed_url.netloc in ('www.youtube.com', 'youtube.com'):
        if parsed_url.path == '/watch':
            query = parse_qs(parsed_url.query)
            return query['v'][0]
        
        if parsed_url.path[:7] == '/embed/':
            return parsed_url.path.split('/')[2]
        
        if parsed_url.path[:3] == '/v/':
            return parsed_url.path.split('/')[2]
    
    return None  # Return None if the URL is not a YouTube URL

def get_transcript(video_id):
    try:
        # Fetch the transcript
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        i = 0
        # Print the transcript with time
        for entry in transcript:
            
            #print(f"Start Time: {entry['start']}, Duration: {entry['duration']}")
            print(f"Text{i}:  {entry['text']}\n")
            i += 1
            
    except Exception as e:
        print(f"An error occurred: {e}")

# Replace 'your_youtube_url_here' with the actual YouTube URL
youtube_url = "https://www.youtube.com/watch?v=vg8xQxuovcU&ab_channel=Theatre101"
video_id = extract_video_id(youtube_url)

if video_id:
    print(f"Extracted Video ID: {video_id}")
    get_transcript(video_id)
else:
    print("Invalid YouTube URL.")

    
    
    
# def extract_video_id(url):
#     parsed_url = urlparse(url)

#     if parsed_url.netloc == 'youtu.be':
#         return parsed_url.path[1:]

#     if parsed_url.netloc in ('www.youtube.com', 'youtube.com'):
#         if parsed_url.path == '/watch':
#             query = parse_qs(parsed_url.query)
#             return query['v'][0]

#         if parsed_url.path[:7] == '/embed/':
#             return parsed_url.path.split('/')[2]

#         if parsed_url.path[:3] == '/v/':
#             return parsed_url.path.split('/')[2]

#     return None  # Return None if the URL is not a YouTube URL

# def get_transcript(video_id):
#     try:
#         # Fetch the transcript
#         transcript = YouTubeTranscriptApi.get_transcript(video_id)
#         transcript_text = ' '.join([f"Start Time: {entry['start']}, Duration: {entry['duration']}, Text: {entry['text']}" for entry in transcript])


#         # Print the transcript with time
# #             for entry in transcript:
# #                 print(f"Start Time: {entry['start']}, Duration: {entry['duration']}")
# #                 print(f"Text: {entry['text']}\n")

#         return transcript_text
#             # Problem is here. It should return a string with the transcript instead of printing this

#     except Exception as e:
#         print(f"An error occurred: {e}")

# def write_to_temp_file_and_load(content):
# # Create a temporary file and write content to it
#     with tempfile.NamedTemporaryFile(delete=True, mode='w+', encoding='utf-8') as temp_file:
#         temp_file.write(content)
#         temp_file.seek(0)  # Move file cursor to the beginning
#         # Read content back from the temporary file while it's still open
#         #read_content = temp_file.read()
#         loader = TextLoader(temp_file.name)
#         document = loader.load_and_split()


#     return document




# youtubevid = 'https://youtu.be/jU0iXjZI2AU'
# video_id = extract_video_id(youtubevid)
# #print(video_id)
# transcripts = get_transcript(video_id)
# type(transcripts)
# documents = write_to_temp_file_and_load(transcripts)


# text_splitter = RecursiveCharacterTextSplitter(
#     # Set a really small chunk size, just to show.
#     chunk_size = 200,
#     chunk_overlap  = 20,
#     length_function = len,
#     is_separator_regex = False,
# )

# docstoembed = text_splitter.split_documents(documents)
# embeddings = OpenAIEmbeddings()


# vectordb = Chroma.from_documents(docstoembed, embedding=embeddings)
#                                  #persist_directory=".")
# retriever = vectordb.as_retriever(search_type="similarity", search_kwargs={"k":2})

# #vectordb.persist()
# memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)
# pdf_qa = ConversationalRetrievalChain.from_llm(OpenAI(temperature=0) , retriever, memory=memory)

# result = pdf_qa({"question": "What is this about?"})


# result["answer"]    
```
